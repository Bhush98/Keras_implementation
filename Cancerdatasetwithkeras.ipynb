```python
# Bulding a neural network to test whether the Cancer is Benign or Melignant from keras

# getting our dependencies 
# Keras is a simple library that is helpful for creating powerfull
# neural networks

from keras.models import Sequential 
from keras.layers import Dense
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
np.random.seed(7)
```

```python
# Getting our dataset i.e our cancer dataset

dataset = pd.read_csv('Downloads/data.csv') 

# Cleaning the data

dataset.dropna() 
dataset.replace('?',-99999,inplace = True)
# we replace them because they are strings and the probability by
# our sigmoid function is in float so it will through an error
# that y is in string should be in float

dataset['diagnosis'].replace('M',1,inplace=True) #Here M represents Melignant
dataset['diagnosis'].replace('B',0,inplace=True) # And B represents Benign
```

```python
# printing our first five instances 
print(dataset.head())
```

```python
# Getting features

X_cancer = dataset[['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean',
                           'compactness_mean','concavity_mean','concave points_mean','texture_worst',
                           'perimeter_worst','area_worst','smoothness_worst','compactness_worst',
                           'concavity_worst','concave points_worst','symmetry_worst',
                           'fractal_dimension_worst']]
```

```python
# Getting target Value

Y_cancer = dataset['diagnosis']
```

```python
# Creating our neural network
# here sequential represents that the layers are one after the other
model = Sequential()
# and dense means that each neuron is connected to every other neuron
# we are using relu activation function in first four layer
model.add(Dense(17,input_dim=17,activation='relu'))
model.add(Dense(16,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(6,activation='relu'))
# and sigmoid in the last layer to get perfect result 
model.add(Dense(1,activation='sigmoid'))
```

```python
# splitting our data in training and testing
X_train,x_test,Y_train,y_test = train_test_split(X_cancer,Y_cancer)
```

```python
# normalizing the data
Scaler = MinMaxScaler().fit(X_train)
X_train = Scaler.transform(X_train)
x_test = Scaler.transform(x_test)
```

```python
# specifying our compile function
# loss function used is binary_crossentropy because in our situation
# we have two conditions 1 or 0 i.e Melignant or benign
# and our optimizer is adam i.e type of gradient descent
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
# fitting our data to our model
# we are running this network for 200 iterations with batch size 10

model.fit(X_train,Y_train,epochs=200,batch_size=10)
# evaluating our model 
score = model.evaluate(x_test,y_test)
# printing our accuracy score
print('Accuracy:%2f'%(score[1]*100))
```
